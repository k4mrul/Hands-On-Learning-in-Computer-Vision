## Production Inference Configuration

This block handles **Production Inference**—executing the trained model on video input while maintaining system stability and producing clean, production-ready outputs.

###  Key Parameters

- **`stream=True`**  
  Processes the video **frame-by-frame** instead of loading it entirely into memory, preventing RAM overload and enabling long-duration video inference.

- **`imgsz=416`**  
  Matches the training resolution to ensure **optimal detection accuracy** and consistent model performance.

- **`retina_masks=True`**  
  Generates **high-resolution, pixel-accurate segmentation masks**, ensuring precise visualization of detected objects (lemons).

- **`show_boxes=False`, `show_labels=False`**  
  Removes bounding boxes and labels to eliminate visual clutter, producing a **clean, industrial-grade output video**.

---

##  Key Benefits

### 1. Stability
Configured with **`workers=0`** and **`device="cpu"`**, ensuring reliable execution on standard machines without GPU dependency or crashes.

### 2. Visual Clarity
Outputs a **“Clean View” video**, highlighting only the segmented objects and their motion—ideal for presentations and analysis.

### 3. Efficiency
Utilizes a **generator-based loop** (`for _ in ...: pass`) to process large volumes of frames efficiently, preventing memory leaks and enabling scalable inference.

```python
from ultralytics import YOLO

MODEL_PATH = r"runs\segment\train\weights\best.pt"
VIDEO_PATH = r"Untitled design (13).mp4"

model = YOLO(MODEL_PATH)

# The arguments show_boxes=False and show_labels=False 
# tell the internal plotter to only save the segmentation masks.
for _ in model.predict(
    source=VIDEO_PATH,
    save=True,
    stream=True,      
    imgsz=416,
    device="cpu",
    workers=0,
    conf=0.2,
    retina_masks=True,
    show_boxes=False,  # Removes the bounding box
    show_labels=False, # Removes the text label and confidence score
    show_conf=False    # Extra safety to ensure no confidence text appears
):
    pass

print("✅ Video processed and saved with masks only (no boxes/labels).")
print("✅ Full video processed without RAM crash")
```

## Calibration Module Breakdown

This script acts as a **Manual ROI (Region of Interest) Calibration Tool**, ensuring the inspection line is precisely aligned based on the actual camera perspective and conveyor position.

### ⚙️ Core Functionality


- **`setMouseCallback`**  
  Enables an interactive listener that captures **mouse click events**. When the user clicks on the conveyor belt, the exact **Y-pixel coordinate** is recorded.

- **Dynamic Visual Feedback**  
  Displays a **temporary yellow horizontal line** along with coordinate text, allowing real-time preview of the selected ROI height before confirmation.

- **`key == 13 (Enter)`**  
  Confirms and locks the selected **`selected_y`** value, then closes the calibration window for direct use in the main counting pipeline.

---

##  Key Benefits

### 1. Precision
Removes manual guesswork by allowing ROI selection directly from the **actual video frame**, ensuring highly accurate placement.

### 2. Adaptability
Quickly recalibrate if the **camera angle shifts** or conveyor alignment changes—no need for hardcoded adjustments.

### 3. Visual Confirmation
Live preview ensures the ROI line is correctly positioned and avoids interference with **mechanical components or irrelevant regions**.

```python
import cv2

VIDEO_PATH = r"Untitled design (13).mp4"
cap = cv2.VideoCapture(VIDEO_PATH)

# Jump to 5 seconds
cap.set(cv2.CAP_PROP_POS_MSEC, 5000)
success, frame = cap.read()

if not success:
    print("Error: Could not read frame at 5 seconds.")
    cap.release()
else:
    selected_y = 0

    def click_event(event, x, y, flags, params):
        global selected_y
        if event == cv2.EVENT_LBUTTONDOWN:
            selected_y = y
            # Draw a temporary line to show where you clicked
            temp_frame = frame.copy()
            cv2.line(temp_frame, (0, y), (frame.shape[1], y), (0, 255, 255), 2)
            cv2.putText(temp_frame, f"Selected Height: {y}px", (10, y - 10), 
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2)
            cv2.imshow("Calibrate Height", temp_frame)
            print(f"Current selection Y = {y}")

    print("--- CALIBRATION MODE ---")
    print("1. Click on the belt to set the inspection line height.")
    print("2. Press 'ENTER' to confirm and close.")
    
    cv2.imshow("Calibrate Height", frame)
    cv2.setMouseCallback("Calibrate Height", click_event)
    
    # Wait for Enter Key (13)
    while True:
        key = cv2.waitKey(0)
        if key == 13: # Enter key
            break

    cv2.destroyAllWindows()
    cap.release()
    print(f"\n✅ CALIBRATION FINISHED")
    print(f"Use this value for line_y: {selected_y}")
```

## ⚙️ Dual-Belt Logic & Tracking Breakdown

This module serves as the **core engine** of the system, combining **persistent tracking, spatial partitioning, and high-speed counting logic** for dual-conveyor environments.

---

##  Core Components

### 1. ByteTrack & Persistent IDs
Utilizes **`persist=True`** with **ByteTrack** to assign a **unique ID** to each detected lemon.  
This ensures **no double-counting**, even if objects are temporarily blurred, occluded, or overlapping.

---

### 2. Dynamic Inspection Masking
Applies a **selective visualization strategy**:
- The region **above the inspection line** remains raw (original video).
- The region **below the line** displays **AI segmentation masks only**.

This creates a clean and professional **“Inspection Zone” view**, focusing attention where counting happens.

---

### 3. Trigger-Zone Logic
Instead of relying on a single pixel line, the system uses a **40-pixel buffer zone** for robust detection.

- **Detection Phase**  
  When a lemon’s center enters the zone  
  (`line_y → line_y + 40`), it triggers a counting event.

- **Validation Phase**  
  A **`counted_ids` set** ensures each tracked object is counted **only once**, eliminating duplicates.

---

### 4. Spatial Partitioning (Belt A vs. Belt B)
Implements **horizontal frame splitting** using `belt_divider`:

- Computes object center (`cx`)
- Instantly assigns the object to:
  - **Left Side → Belt A**
  - **Right Side → Belt B**

This enables **simultaneous, independent counting** across parallel conveyor streams.

---

##  Performance Highlights

### 1. Stable Video Output
Uses **XVID codec** to generate reliable, **uncorrupted `.avi` files** suitable for industrial review and analysis.

### 2. Centered UI Design
Dynamically aligns labels to the center, creating a **balanced and dashboard-like visualization**.

### 3. Trigger Buffer Reliability
The **buffer-based trigger zone** ensures **near 100% counting accuracy**, even under **frame drops or variable FPS conditions**.

```python
import cv2
from ultralytics import YOLO
import os

# 1. Setup
MODEL_PATH = r"runs\segment\train\weights\best.pt"
VIDEO_PATH = r"Untitled design (13).mp4"
OUTPUT_FILE = "Lemon_Count_Final.avi" 

model = YOLO(MODEL_PATH)
cap = cv2.VideoCapture(VIDEO_PATH)

width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps    = int(cap.get(cv2.CAP_PROP_FPS))

fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(OUTPUT_FILE, fourcc, fps, (width, height))

# --- CONFIGURATION ---
line_y = 550
trigger_buffer = 40 
belt_divider = width // 2

counted_ids = set()
count_A = 0
count_B = 0

try:
    while cap.isOpened():
        success, frame = cap.read()
        if not success:
            break

        # Inference on full frame for stable tracking
        results = model.track(frame, persist=True, tracker="bytetrack.yaml", 
                              conf=0.3, imgsz=640, verbose=False)
        
        display_frame = frame.copy()

        if results[0].boxes.id is not None:
            boxes = results[0].boxes.xyxy.cpu().numpy()
            ids = results[0].boxes.id.int().cpu().numpy()
            
            # Draw masks globally, but we will "clean" the top part later
            if results[0].masks is not None:
                annotated_frame = results[0].plot(boxes=False, labels=False, conf=False)
                # Keep raw video ABOVE the line
                annotated_frame[0:line_y, 0:width] = frame[0:line_y, 0:width]
                display_frame = annotated_frame

            for box, track_id in zip(boxes, ids):
                cx = int((box[0] + box[2]) / 2)
                cy = int((box[1] + box[3]) / 2)
                
                # --- TRIGGER ZONE LOGIC ---
                # 1. Is the lemon ID new?
                # 2. Is the center point below the line (418)?
                # 3. Is the center point within the buffer zone (418 to 448)?
                if track_id not in counted_ids:
                    if line_y < cy < (line_y + trigger_buffer):
                        if cx < belt_divider:
                            count_A += 1
                        else:
                            count_B += 1
                        counted_ids.add(track_id)

        # UI Drawing
        cv2.rectangle(display_frame, (0, 0), (width, 60), (30, 30, 30), -1)
        
        # Centered Alignment
        text_A = f"BELT A: {count_A}"
        text_B = f"BELT B: {count_B}"
        (twA, _), _ = cv2.getTextSize(text_A, cv2.FONT_HERSHEY_DUPLEX, 1, 2)
        
        cv2.putText(display_frame, text_A, (belt_divider - twA - 20, 42), cv2.FONT_HERSHEY_DUPLEX, 1, (0, 255, 0), 2)
        cv2.putText(display_frame, text_B, (belt_divider + 20, 42), cv2.FONT_HERSHEY_DUPLEX, 1, (255, 150, 0), 2)
        
        # Draw the Inspection Line
        cv2.line(display_frame, (0, line_y), (width, line_y), (255, 255, 255), 2)

        out.write(display_frame)
        cv2.imshow("Production Counter", display_frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

finally:
    cap.release()
    out.release()
    cv2.destroyAllWindows()
    print(f"✅ Final Counts -> Belt A: {count_A} | Belt B: {count_B}")
```
